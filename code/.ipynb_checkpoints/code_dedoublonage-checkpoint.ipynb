# Objectif de l'étude : 
## A partir d'un répertoire donné, identifier les fichiers ayant des contenus identiques.
## Ces fichiers pouvant être de tous types (Word, Excel, pdf, csv, txt, json, ...)

# Méthode de résolution :
### 1- A l'aide d'une fonction de hachage type md5, calcluer le hash de chaque fichier du répertoire cible.
### 2 - Stocker les résultats dans un dictionnaire type {hash : nom_fichier}.
### 3 - Créer un dictionnaire ou une liste vide, et y stocker les fichiers n'ayant pas leur hash déjà présent.

```python
from os import listdir
from os.path import isfile, join
file_path = ("C:/Users/Patrick Menan/Desktop/projet/dedoublonnage/data/Buffer-in")
liste_of_file = [file for file in listdir(file_path) if isfile(join(file_path, file))]
```

```python
#liste de tous les fichiers présent dans "liste_of_file"
liste_of_file
```

```python
#nombre de fichier dans le repertoire
len(liste_of_file)
```

```python
#fonction de hachage
import hashlib
def hash_file(file):
    return hashlib.md5(open(file, 'rb').read()).hexdigest()
```

```python
#calcul du hash de tous les fichier du repertoire cible.
#resultat est stocké dans un objet dictoinnaire
resultat = {hash_file(join(file_path, file)) : file for file in listdir(file_path) if isfile(join(file_path, file))}
```

```python
#resultat de la fonction de hachage
resultat
```

```python
#création de deux dictionnaires. 
#my_dict => contient les fichiers principaux
#my_dict_doublon => contient les fichiers pouvant être en doublon triplon

my_dict = {}
my_dict_doublon = {}
for file in liste_of_file:
    if hash_file(join(file_path, file)) not in my_dict.keys():
        my_dict[hash_file(join(file_path, file))] = file
    else :
        print("le fichier "+file+" est peut-être en double\n")
        my_dict_doublon[hash_file(join(file_path, file))] = file   
```

```python
#identification claie du lien entre le fichier principal et fichier ayant potentiellement le même contenu.
for key in my_dict_doublon.keys():
    if key in my_dict.keys():
        print ("le fichier "+ my_dict_doublon[key]+" et "+my_dict[key]+" ont peut-être des contenus identiques\n")
```

```python
#fonction générale d'identification des doublons 
import pandas as pd
def dedoublonage (file_path): 
    liste_of_file = [file for file in listdir(file_path) if isfile(join(file_path, file))]
    my_dict = {}
    my_dict_doublon = {}
    for file in liste_of_file: 
        if hash_file(join(file_path, file)) not in my_dict.keys(): 
            my_dict[hash_file(join(file_path, file))] = file
        else : 
            my_dict_doublon[hash_file(join(file_path, file))] = file 
            #os.remove(join(file_path, file))
    
    df = pd.DataFrame({
        'hash_code' : [key for key in my_dict.keys() if key in my_dict_doublon.keys()], 
        'principale_file_name' : [my_dict[key] for key in my_dict.keys() if key in my_dict_doublon.keys()],
        'double_file_name' : [my_dict_doublon[key] for key in my_dict.keys() if key in my_dict_doublon.keys()]
    })
    
    """
    decommenter la ligne #os.remove() si on souhaite supprimer les fichiers pouvant être en double.
    il est recommander de faire une analyse préalable afin de vérifier que les fichiers 
    ont réellement des contenus identique.
    """
    
    return df
    
```
